```python
#Extracting dependent column from independent columns.
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#Train , Test & Split for Model Building
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30, random_state=1)
```

```python
X_train #Train Data including Independent Features
```

```python
X_train.shape #shape of train data including independent features
```

```python
X_test #Test Data including Independent Features
```

```python
X_test.shape #shape of test data including independent features
```

```python
y_train # Train Data with Target Feature
```

```python
y_train.shape #shape of train data including target feature
```

```python
y_test # Test Data with Target Feature
```

```python
y_test.shape #shape of test data including target feature
```

```python
#Standardize the dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train # train data after scaling
```

```python
X_test # test data after scaling
```

#Note - As we use gradient desecent to achieve  global minima in linear regression model so thats why we have use standard scaler to standardize the units of all the independent features to make the units of all independent features same.

```python
#Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#print the coeffcients and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameters the model has been trained
regression.get_params()
```

```python
# prediction on test data
```

```python
reg_test_pred=regression.predict(X_test)
```

```python
reg_test_pred
```

**#Assumptions**

```python
# plot a scatte rplot with test data
```

```python
plt.scatter(y_test,reg_test_pred)
```

```python
#Note - if the plot comes out to be linear it means model is predicting good.
```

```python
#Residuals
residuals= y_test-reg_test_pred
residuals
```

```python
#plot of residuals
sns.displot(residuals,kind="kde")
```

```python
#Note - errors are also almost normal distribution.
```

```python
# scatter plot wrt prediction and residuals
plt.scatter(reg_test_pred,residuals)
```

```python
#Note - evenly distributed , but most the data points are uniformly distributed.
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_test_pred))
print(mean_squared_error(y_test,reg_test_pred))
print(np.sqrt(mean_squared_error(y_test,reg_test_pred)))
```

```python
# r-square and Adjusted-r-square.
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,reg_test_pred)
print(score)
```

```python
#Adjusted R2- Square 
#R2 = 1-[(1-R2)*(N-1)/(N-K-1)]
```

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
